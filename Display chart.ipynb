## Display graph overview

```python
import pickle

# File paths
GRAPH_FILE = 'user_graph.pkl'
VISITED_FILE = 'visited_users.pkl'
QUEUE_FILE = 'user_queue.pkl'

# Load and display the graph data
print("=== Graph Data ===")
with open(GRAPH_FILE, 'rb') as f:
    graph = pickle.load(f)

# Check if the graph is not empty
if graph:
    for user, data in graph.items():
        print(f"User: {user}")
        print(f"  Followers Count: {data.get('followers_count', 'N/A')}")
        print(f"  Connections: {data.get('connections', [])}")
        print("-" * 40)
else:
    print("Graph is empty.")

# Load and display the visited users
print("\n=== Visited Users ===")
with open(VISITED_FILE, 'rb') as f:
    visited = pickle.load(f)

if visited:
    print(f"Total Visited Users: {len(visited)}")
    print("Visited Users List:")
    for user in visited:
        print(user)
else:
    print("No users have been visited yet.")

# Load and display the queue
print("\n=== User Queue ===")
with open(QUEUE_FILE, 'rb') as f:
    queue = pickle.load(f)

if queue:
    print(f"Users in Queue: {len(queue)}")
    print("Queue Contents:")
    for item in queue:
        print(item)
else:
    print("Queue is empty.")
```
